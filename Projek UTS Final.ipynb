```python
import pandas as pd
```

```python
dataset1 = pd.read_csv("MQTT DDoS Publish Flood.csv")
```

```python
dataset1
```

```python
dataset2 = pd.read_csv("MQTT DoS Connect Flood.csv")
```

```python
dataset2
```

```python
dataset3= pd.read_csv("MQTT DoS Publish Flood.csv")
```

```python
dataset3
```

```python
hasilgabung =pd.concat([dataset1,dataset2,dataset3],ignore_index=True)
```

```python
hasilgabung.columns.values
```

```python
a = hasilgabung.iloc[:,7: 76]
```

```python
a
```

```python
b = hasilgabung.iloc[:,83: 84]
```

```python
b
```

```python
from sklearn.model_selection import train_test_split
```

```python
a_train,a_test,b_train,b_test = train_test_split(a,b, test_size = 0.2, random_state = 42)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
zah = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
zah.fit(a_train,b_train)
```

```python
b_pred = zah.predict(a_test)
```

```python
b_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(b_test,b_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(zah, feature_names = a.columns.values, class_names = np.array([ 'MQTT DDoS Publish Flood','MQTT DoS Connect Flood','MQTT DoS Publish Flood']), filled = True)
plt.show()
```

```python
import seaborn as lol
from sklearn import metrics
label = np.array([ 'MQTT DDoS Publish Flood','MQTT DoS Connect Flood','MQTT DoS Publish Flood'])
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(b_test, b_pred)
plt.figure(figsize=(10, 10))
lol.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
